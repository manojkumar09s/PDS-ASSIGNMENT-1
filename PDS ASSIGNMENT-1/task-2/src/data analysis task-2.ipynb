```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset

import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/PDS ASSIGNMENT/StudentsPerformance.csv')
```

```python
df.tail()
```

```python
# Task 1: Pie chart showing the distribution of genders
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Genders')
plt.show()
```

```python
# Task 2: Bar chart showing the distribution of parental level of education
edu_counts = df['parental level of education'].value_counts()
plt.figure(figsize=(10, 6))
edu_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Parental Level of Education')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Task 3: Stacked bar chart showing lunch type and test completion
lunch_completion_counts = df.groupby(['lunch', 'test preparation course']).size().unstack()
lunch_completion_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], figsize=(10, 6))
plt.title('Test Completion based on Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Test Completion')
plt.tight_layout()
plt.show()
```

```python

# Task 4: Box plot showing distribution of test scores by gender
plt.figure(figsize=(10, 6))
plt.title('Distribution of Test Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Test Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.boxplot([df[df['gender'] == 'female'].index, df[df['gender'] == 'male'].index])
plt.xticks([1, 2], ['Female', 'Male'])
plt.tight_layout()
plt.show()
```

```python
# Task 5: Violin plot showing distribution of test scores by parental level of education
plt.figure(figsize=(12, 8))
plt.title('Distribution of Test Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Test Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.violinplot([df[df['parental level of education'] == edu].index for edu in df['parental level of education'].unique()], showmeans=False)
plt.xticks(range(1, len(df['parental level of education'].unique()) + 1), df['parental level of education'].unique(), rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Pie chart showing the distribution of test completion

completion_counts = df['test preparation course'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(completion_counts, labels=completion_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Test Completion')
plt.show()
```

```python
# Box plot of test scores by lunch type
plt.figure(figsize=(10, 6))
plt.title('Distribution of Test Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Test Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.boxplot([df[df['lunch'] == 'standard'].index, df[df['lunch'] == 'free/reduced'].index])
plt.xticks([1, 2], ['Standard', 'Free/Reduced'])
plt.tight_layout()
plt.show()
```
