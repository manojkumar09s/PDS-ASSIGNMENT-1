```python
# Load the dataset

import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/PDS ASSIGNMENT/task1_data.csv')
```

```python
df
```

```python
frailty_counts = df['Frailty'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(frailty_counts, labels=frailty_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Frailty')
plt.show()
```

```python
# Bar chart showing the count of individuals in each Age group
age_bins = pd.cut(df['Age'], bins=[0, 20, 40, 60, 80], labels=['0-20', '21-40', '41-60', '61-80'])
age_group_counts = age_bins.value_counts().sort_index()
age_group_counts.plot(kind='bar', color='skyblue', figsize=(8, 6))
plt.title('Count of Individuals in Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
